```python
# Custom Data에 맞는 YAML 파일 만들기
data_yaml_path = "/content/drive/MyDrive/비타민-2024-2-CV/Insideout_character_segmentation.v1i.yolov8/data.yaml"
data = {
    "train": "/content/drive/MyDrive/비타민-2024-2-CV/Insideout_character_segmentation.v1i.yolov8/train",
    "val": "/content/drive/MyDrive/비타민-2024-2-CV/Insideout_character_segmentation.v1i.yolov8/valid",
    "test": "/content/drive/MyDrive/비타민-2024-2-CV/Insideout_character_segmentation.v1i.yolov8/test",
    "nc": 3,
    "names": ['Anger', 'Joy', 'Sadness']
}

# dataset 에 맞게 yaml 파일 생성
with open(data_yaml_path, "w") as f:
    yaml.dump(data, f)
```

```python
from ultralytics import YOLO

# Google Drive에 저장된 모델 파일 경로
model_path = "/content/drive/MyDrive/Models/yolov8n-seg.pt"

# 모델 로드
model = YOLO("yolov8n-seg.pt")
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

```python
# 학습 시작 (runs 폴더를 Google Drive 안에 저장하도록 설정)
model.train(data=data_yaml_path, epochs=300, patience=30, batch=32, imgsz=416, project="/content/drive/MyDrive/비타민-2024-2-CV/runs", name="experiment")
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

```python
results = model.predict(source="/content/drive/MyDrive/BITAmin/Project/Datasets/test/images", save=True)
```
